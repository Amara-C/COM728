```python
import csv
import os

file_name = "./Airbnb_UK_2022.csv"

 
#error handling with the try except block

try:
    #open csv file
    with open (file_name) as csv_file:
    #read the csv file
        csv_reader = csv.reader(csv_file)
    #read in heading and store seperately
        heading = next(csv_file)
    
    #code to display headings
        print (heading)
    
except IOError:
    print (f"sorry, the file {file_name} cannot be found")

```

```python
#  ask user to enter the host id
host_id = input("Please enter the host ID you would like to retrieve: ")

# including the try except block to handle errors
try:
    with open(file_name, 'r', encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file)
        column_data = []
        for row in csv_reader:
            if row[0] == host_id:
                row_data = [row[1], row[3], row[2], row[5], row[4]]
                column_data.append(row_data)
        print(column_data)
except FileNotFoundError:
    print(f"invalid! the id '{host_id}' cannot be found, please enter the correct host id.")
except Exception as e:
    print(f"An error occurred while processing the file: {e}")
```
